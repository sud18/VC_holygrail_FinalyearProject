```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn import linear_model, datasets
```

```python
df = pd.read_csv('final_data.csv', encoding = "ISO-8859-1")
```

```python
sub_df = df.iloc[:,1:19]
```

# Try logreg, rf, perceptron, SVM...

```python
X = result_df
Y = y_df
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=100)
print ('Number of samples in training data:',len(x_train))
print ('Number of samples in validation data:',len(x_test))
```

```python
# RF

from sklearn.ensemble import RandomForestClassifier

## Initialize a Random Forest classifier with 50 trees
# hyperparameter n_estimators always set in instantiation

forest = RandomForestClassifier(n_estimators = 50)
```

```python
%%time
# Fit the forest to the training set, using the bag of words as 
# features and the sentiment labels as the target variable

forest = forest.fit(x_train, y_train) # can take 2 mins to run
```

```python
#Make predictions

train_predictions = forest.predict(x_train)
valid_predictions = forest.predict(x_test)
```

```python
# Accuracy

metrics.accuracy_score(y_train,train_predictions)
```

```python
metrics.accuracy_score(y_test,valid_predictions)
```

```python
# Confusion matrix
# Is the number of False Positives and True negatives approx 50/50?
metrics.confusion_matrix(y_test,valid_predictions)
```

```python
importances = forest.feature_importances_
# returns relative importance of all features.
# they are in the order of the columns
print(importances)
```

```python
# Logreg

# Name our regression object
logreg = linear_model.LogisticRegression(C=1e5)

# we create an instance of Neighbours Classifier and fit the data.
logreg.fit(x_train, y_train)
```

```python
train_predictions_lg = logreg.predict(x_train)
valid_predictions_lg = logreg.predict(x_test)
```

```python
# Accuracy

metrics.accuracy_score(y_train,train_predictions_lg)
```

```python
metrics.accuracy_score(y_test,valid_predictions_lg)
```

```python
# Confusion matrix
# Is the number of False Positives and True negatives approx 50/50?
metrics.confusion_matrix(y_test,valid_predictions_lg)
```

```python
# SVM

from sklearn.svm import SVC

model = SVC()

model.fit(x_train, y_train)
```

```python
train_predictions_svm = model.predict(x_train)
valid_predictions_svm = model.predict(x_test)
```

```python
# Accuracy

metrics.accuracy_score(y_train,train_predictions_svm)
```

```python
metrics.accuracy_score(y_test,valid_predictions_svm)
```

```python
model.score(x_train, y_train)
#Predict Output
predicted= model.predict(x_test)
```
